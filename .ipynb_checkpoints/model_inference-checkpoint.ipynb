# Этот ноутбук был создан для inference. Есть две опции:
## 1) Загрузить картинку с компьютера
## 2) Протестировать снимок с веб-камеры

```python
import cv2
from time import sleep
import numpy as np
import os
from matplotlib import pyplot as plt

import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model
```

```python
# Загружаем веса модели
model = load_model('mask_detection_model.h5')
```

```python
# Это функция для проверки модели на конкретном файле, к нему нужно указать путь.
def inference_from_file(path, model = model):
    # загружаем и обрабатываем картинку
    test_image = load_img(path, target_size = (128, 128), color_mode = 'rgb')
    test_image = img_to_array(test_image)/255
    plt.imshow(test_image)
    plt.axis('off')
    test_image = np.expand_dims(test_image, axis = 0)
    
    prediction = float(np.squeeze(model.predict(test_image)))
    if prediction < 0.5:
        print('Model thinks there is no mask! Output: {}'.format(round(prediction, 3)))
    else:
        print('Model thinks there is a mask! Output: {}'.format(round(prediction, 3)))
        
# Это функция для проверки модели по снимку веб-камеры
def inference_webcam(model = model):
    webcam = cv2.VideoCapture(0)
    webcam.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
    webcam.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)
    print('Snapshot in 3 seconds!')
    sleep(1)
    print('Snapshot in 2 seconds!')
    sleep(1)
    print('Snapshot in 1 second!')
    sleep(1)
    check, frame = webcam.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame, (128, 128))
    frame = np.array(frame)/255
    plt.imshow(frame)
    plt.axis('off')
    webcam.release()
    frame = np.expand_dims(frame, axis = 0)
    
    prediction = float(np.squeeze(model.predict(frame)))
    if prediction < 0.5:
        print('Model thinks there is no mask! Output: {}'.format(round(prediction, 3)))
    else:
        print('Model thinks there is a mask! Output: {}'.format(round(prediction, 3)))
```

```python
# Эта ячейка для первой опции
path_to_photo = 'Test_3.jpg'
inference_from_file(path_to_photo)
```

```python
# Это ячейка для второй опции
inference_webcam()
```
